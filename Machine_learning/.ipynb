```python
df = pd.read_csv('/Users/miriambenali/Desktop/Project-Simplon/RetentionAI/data/dt.csv')
```

```python
num_cols = df.select_dtypes(include='number').columns
print(num_cols)
df[num_cols]= df[num_cols].fillna(df[num_cols].mean())
```

```python
cat_cols= df.select_dtypes(include='object').columns
print(cat_cols)
for col in cat_cols:
    df[col].fillna(df[col].mode()[0], inplace=True)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

```

```python
fig, axe =plt.subplots(1,8, figsize=(18,5))

sns.countplot(x='BusinessTravel',hue='Attrition', data=df, ax=axe[0])

sns.countplot(x='Department',hue='Attrition', data=df, ax=axe[1])

sns.countplot(x='EducationField',hue='Attrition', data=df, ax=axe[2])

sns.countplot(x='Gender',hue='Attrition', data=df, ax=axe[3])

sns.countplot(x='JobRole',hue='Attrition', data=df, ax=axe[4])

sns.countplot(x='MaritalStatus',hue='Attrition', data=df, ax=axe[5])

sns.countplot(x='Over18',hue='Attrition', data=df, ax=axe[6])

sns.countplot(x='OverTime', hue='Attrition', data=df)

plt.tight_layout()
plt.show()
```

```python
# On doit encoder 'Attrition' pour qu'on puisse faire la correlation

df['Attrition_num'] = df['Attrition'].map({'No': 0, 'Yes': 1})
```

```python
#verifier si on a des colums vide
df.isna().all

#on constate qu'il ya 3 columns manqués
```

```python
#afficher les noms des columns vides

num_cols = df.select_dtypes(include='number')
empty_cols = [col for col in num_cols if df[col].nunique() <= 1 or df[col].isna().all()]
print("Colonnes vides ou constantes :", empty_cols)
```

```python
df = df.drop(columns=empty_cols)
```

```python
# Sélection des colonnes numériques
num_cols = df.select_dtypes(include='number')

# matrice de correlation
corr_matrix = num_cols.corr()

plt.figure(figsize=(14,10))
sns.heatmap(corr_matrix, annot=True, fmt=".2f", cmap='coolwarm')
plt.title('Matrice de correlation')
plt.show()

```
